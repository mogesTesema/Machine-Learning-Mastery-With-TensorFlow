 **The dot product**
 Matrix multiplication s also refered to as the dot product you can perform matrix mulitplication using:
 * `tf.matmul()`
 * `tf.tensordot()`


```python
# perform the dot product on X and Y (requires X or Y to be transpose)
# import numpy as np
X = np.arange(1,7, dtype=np.int32)
Y = np.arange(7,13, dtype=np.int32)
X = tf.constant(X, shape=(3,2))
Y = tf.constant(Y, shape = (3,2))
tf.tensordot(tf.transpose(X),Y, [[0],[1]]), tf.matmul(X,Y,True)
```

```python
# Perform matrix multiplication between X and Y (transposed)
tf.matmul(X, tf.transpose(Y))
```

```python
# Perfomr matrix multiplication between X and Y ( Reshaped)
tf.matmul(X, tf.reshape(Y, shape = (2,3)))
```

```python
# check the values of Y, reshape Y and transposed Y
print("Normal Y:")
print(Y,"\n")

print("Y reshpaed to (2,3):")
print(tf.reshape(Y,(2,3)),"\n")
print("Transposed Y: ")
print(tf.transpose(Y))
```

### Let's go through the following forms of aggregation
* Ge the minimum
* get the maximun
* get the mean fo a tensor
* get the sum of a tensor

```python
aggregated_tensor = tf.constant(np.random.randint(-10, 50, size = 27),shape = (3,3,3))
max_value = tf.reduce_max(aggregated_tensor)
min_value = tf.math.reduce_min(aggregated_tensor)
sum_value = tf.math.reduce_sum(aggregated_tensor)
average_value = tf.math.reduce_mean(aggregated_tensor)

print(f"Data of the aggregated_tensor aggregation value.{aggregated_tensor.numpy}\nnumber of elements: {tf.size(aggregated_tensor)}")
print("max value: ",max_value.numpy())
print("min value: ", min_value.numpy())
print("average value: ", average_value.numpy())
print("Total sum: ", sum_value.numpy())
```

⛑ **Exercise:** Find variance and standard divation of aggregated_tensor

```python
variance = tf.math.reduce_variance(tf.cast(aggregated_tensor,dtype=tf.float32))
standard_divation = tf.math.reduce_std(tf.cast(aggregated_tensor,dtype=tf.float32))
variance.numpy(), standard_divation.numpy()
```

 ### find the positional maximum and minimum

```python
# create a new tensor for finding positional mininum and maximum
tf.random.set_seed(21)
F = tf.random.uniform(shape = [50], seed = 21)
#Find the positional maximum
max_value_index = tf.argmax(F)
max_value_index
#Index on our largest value position
F[tf.argmax(F)]
# check for equality
F[max_value_index] == tf.math.reduce_max(F)
# Find the positional minimum
min_value_index = tf.argmin(F)
# check the equality
tf.math.reduce_min(F) == F[min_value_index]


```
