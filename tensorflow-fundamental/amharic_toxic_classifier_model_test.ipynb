```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
model_name = "amengemeda/amharic-hate-speech-detection-mBERT"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
test_text = [
    "ይህ በጣም የሚያምር ጽሁፍ ነው!",
    "በጣም መጥፎ እና አሳፋሪ ቃል ነው!",
    "በጣም ደስተኛ ነኝ"
]

encoded_text = tokenizer(test_text, padding=True, truncation=True,return_tensors='pt')
```

```python
import torch
import tensorflow as tf
with torch.no_grad():
  result = model(**encoded_text)
print(result)
```

```python
tf.shape(result.logits)
predictions = torch.argmax(result.logits,dim=1)
predictions
```

```python
id2label = model.config.id2label

print("Input Text | Predicted Class")
print("------------------------------")
for i, text in enumerate(test_text):
    predicted_class_id = predictions[i].item()
    predicted_label = id2label[predicted_class_id] # Use the id2label dictionary to get the string
    print(f"'{text}' | {'positive' if predicted_label == 'LABEL_0' else 'negative'}")
```
